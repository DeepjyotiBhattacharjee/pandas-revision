# scaling

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/sharmaroshan/Social-Networks-Ads/master/Social_Network_Ads.csv")
df.head()
```

```python
df = df.iloc[:,2:]
df
```

```python
X = df.drop('Purchased',axis=1)
X
```

```python
y = df['Purchased']
y
```

### train test split

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.25)
```

```python
X_train
```

```python
y_train
```

```python
X.shape
```

```python
X_train.shape
```

## standard scaler

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaled_X_train = scaler.fit_transform(X_train)
scaled_X_train
```

```python
scaled_X_test = scaler.transform(X_test)
scaled_X_test
```

```python
X_train_scaled = pd.DataFrame(scaled_X_train,columns = X_train.columns)
X_train_scaled
```

```python
X_train.describe().T
```

```python
X_train_scaled.describe().T
```

```python
import numpy as np
np.round(X_train_scaled.describe().T,1)
```

### effect of scaling on data

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12,5))
ax1.scatter(X_train['Age'],X_train['EstimatedSalary'])
ax1.set_title("Before scaling")

ax2.scatter(X_train_scaled['Age'],X_train_scaled['EstimatedSalary'],color='red')
ax2.set_title("After scaling")

plt.show()
```

```python
import seaborn as sns
sns.kdeplot(data = X_train['EstimatedSalary'])
```

```python
sns.kdeplot(data = X_train_scaled['EstimatedSalary'])
```

### ordinalencoder - used on ordinal data

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.25)
X.shape
```

```python
X_train.shape
```

# handle missing values

```python
df3 = pd.read_csv("https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv")
df3.head()
```

```python
df3 = df3[['Age','Fare','SibSp','Survived']]
df3
```

```python
df3.info()
```

```python
df3.isnull().sum()
```

```python
df3.isnull().mean() # age column has 19 percent missing values
```

```python
# numerical col -> mean / median
# categorical col -> mode
# this is simpleimputer
```

```python
X_train, X_test = train_test_split(df3,test_size=0.25)
```

```python
X_train
```

```python
X_test
```

```python
mean_age = X_train['Age'].mean()
mean_age
```

```python
X_train['Age'].fillna(mean_age)
```

```python
X_train['Age'].fillna(mean_age).isnull().sum()
```

```python
# if we have outliers use median
```

```python
df3['Cabin'].mode()
```

```python
df3['Cabin'].fillna(df3['Cabin'].mode()[0])
```
