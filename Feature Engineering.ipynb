```python
df = pd.read_csv("https://raw.githubusercontent.com/campusx-official/100-days-of-machine-learning/main/day42-outlier-removal-using-zscore/placement.csv")
df.head()
```

### if data is normally distributed, we remove the data > 3*std deviation --> Empirical rule

```python
df['cgpa'].std()
```

```python
df['cgpa'].min()
```

```python
df['cgpa'].max()
```

```python
upper_bound = df['cgpa'].mean() + 3*df['cgpa'].std()
lower_bound = df['cgpa'].mean() - 3*df['cgpa'].std()
```

```python
upper_bound
```

```python
lower_bound
```

### values greater than upper bound and less than lower_bound are considered as outliers and can be trimmed (or removed).

```python
df[df['cgpa']>upper_bound] # this data can be dropped
```

```python
df[df['cgpa']<lower_bound] # this data can be dropped
```

```python
df[(df['cgpa']> upper_bound ) | (df['cgpa']<lower_bound)] # this data can be dropped # these are outliers
```

### let's trim these outliers

```python
new_df = df[(df['cgpa']< upper_bound ) & (df['cgpa']>lower_bound)]
new_df
```

```python
sns.distplot(new_df['cgpa'])
```

### capping method -> if value > upper limit then value = upper limit ; if value < lower limit then value = lower limit

```python
upper_bound
```

```python
lower_bound
```

```python
df['cgpa']>upper_bound
```

```python
import numpy as np
np.where(df['cgpa']>upper_bound,upper_bound,np.where(df['cgpa']<lower_bound,lower_bound,df['cgpa']))
```

```python
np.where(df['cgpa']>upper_bound,upper_bound,df['cgpa'])
```

```python
np.where(df['cgpa']<lower_bound,lower_bound,df['cgpa'])
```

```python
df['cgpa'] = np.where(df['cgpa']>upper_bound,upper_bound,np.where(df['cgpa']<lower_bound,lower_bound,df['cgpa']))
```

### IQR on the skewed data

```python
df['placement_exam_marks']
```

```python
df['placement_exam_marks'].skew()
```

```python
sns.boxplot(df['placement_exam_marks'])
```

```python
percentile_25 = df['placement_exam_marks'].quantile(0.25)
percentile_75 = df['placement_exam_marks'].quantile(0.75)
```

```python
percentile_25 # this is Q1
```

```python
percentile_75 # this is Q3
```

```python
iqr = percentile_75 - percentile_25
iqr
```

```python
lower_fence = percentile_25 - (1.5*iqr)
upper_fence = percentile_75 + (1.5*iqr)
```

```python
lower_fence
```

```python
upper_fence
```

```python
df[df['placement_exam_marks']>upper_fence] # these are outliers
```

```python
df[df['placement_exam_marks']<lower_fence] # also outliers
```

```python
df[(df['placement_exam_marks']>upper_fence) | (df['placement_exam_marks']<lower_fence)] # these are outliers
```

```python
# let's remove them
```

```python
new_df = df[(df['placement_exam_marks']<=upper_fence) & (df['placement_exam_marks']>=lower_fence)]
new_df
```

```python
sns.boxplot(new_df['placement_exam_marks'])
```

```python
plt.figure(figsize=(16,10))

plt.subplot(2,2,1)
sns.distplot(df['placement_exam_marks'])

plt.subplot(2,2,2)
sns.boxplot(df['placement_exam_marks'])

plt.subplot(2,2,3)
sns.distplot(new_df['placement_exam_marks'])

plt.subplot(2,2,4)
sns.boxplot(new_df['placement_exam_marks'])

plt.show()
```
