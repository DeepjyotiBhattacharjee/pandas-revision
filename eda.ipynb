```python
df[df['Survived']==0]['Age'].min() # the min age of a person that did not survive was 1
```

```python
df[df['Survived']==1]['Age'].max() # the max age of a person that survived was 80
```

```python
df[df['Survived']==1]['Age'].min() # the min age of a person that survived was 0.42
```
